```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify=y ,random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
ohe_pipeline= Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ohe_hot', OneHotEncoder( handle_unknown= 'ignore', sparse_output=False))
])
```

```python
col_trans= ColumnTransformer(transformers=[
    ('impute', SI, ['Age']),
    ('ohe_pipeline', ohe_pipeline, ohe_cols),
    ('passthrough','passthrough',['Pclass','Age','SibSp','Parch','Fare','Sex_female','Sex_male','Embarked_C','Embarked_Q','Embarked_S']) 
     ],
     remainder='drop',
     n_jobs=-1)
```

```python
# All models fitted in this dataset
```

```python
rfc= RandomForestClassifier()
```

```python
param_grid={
    'n_estimators':[100, 150, 200],
    'min_samples_split':[5, 10, 15],
    'max_depth':[8, 9, 10, 15, 20],
    'min_samples_leaf':[1, 2, 4],
    'criterion':['gini', 'entropy']
}
```

```python
CV_rfc= GridSearchCV(estimator=rfc, param_grid=param_grid, cv=StratifiedKFold(n_splits=5))
```

```python
pipefinalrfc=make_pipeline(col_trans, CV_rfc)
pipefinalrfc.fit(x_train, y_train)
```

```python
print(CV_rfc.best_params)
print(CV_rfc.best_score)
```
